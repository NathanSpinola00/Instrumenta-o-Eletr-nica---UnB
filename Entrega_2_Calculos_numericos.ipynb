```python
# Valores dos ganhos e offsets
Ganho_X = 1.0
Ganho_Y = 1.0
Ganho_Z = 1.02
A_offsetX = -0.04
A_offsetY = -0.04
A_offsetZ = -0.04

# Clculo dos valores de sada usando a equao A_saida = Ganho * A_ref + A_offset
A_saida_X_positive = [Ganho_X * val + A_offsetX for val in Accel_X_pos]
A_saida_X_negative = [Ganho_X * val + A_offsetX for val in Accel_X_neg]
A_saida_Y_positive = [Ganho_Y * val + A_offsetY for val in Accel_Y_pos]
A_saida_Y_negative = [Ganho_Y * val + A_offsetY for val in Accel_Y_neg]
A_saida_Z_positive = [Ganho_Z * val + A_offsetZ for val in Accel_Z_pos]
A_saida_Z_negative = [Ganho_Z * val + A_offsetZ for val in Accel_Z_neg]


# Plotando os grficos com as sadas sem compensao e com compensao
plt.figure(figsize=(15, 10))

plt.subplot(3, 2, 1)
plt.plot(amostra, Accel_X_pos, label='Accel +X Original', linestyle='-', color='blue')
plt.plot(amostra, A_saida_X_positive, label='Accel +X Compensado', linestyle='--', color='red')
plt.title('Accel +X')
plt.xlabel('Amostra')
plt.ylabel('Valor de Acelerao')
plt.legend()

plt.subplot(3, 2, 2)
plt.plot(amostra, Accel_X_neg, label='Accel -X Original', linestyle='-', color='blue')
plt.plot(amostra, A_saida_X_negative, label='Accel -X Compensado', linestyle='--', color='red')
plt.title('Accel -X')
plt.xlabel('Amostra')
plt.ylabel('Valor de Acelerao')
plt.legend()

plt.subplot(3, 2, 3)
plt.plot(amostra, Accel_Y_pos, label='Accel +Y Original', linestyle='-', color='blue')
plt.plot(amostra, A_saida_Y_positive, label='Accel +Y Compensado', linestyle='--', color='red')
plt.title('Accel +Y')
plt.xlabel('Amostra')
plt.ylabel('Valor de Acelerao')
plt.legend()

plt.subplot(3, 2, 4)
plt.plot(amostra, Accel_Y_neg, label='Accel -Y Original', linestyle='-', color='blue')
plt.plot(amostra, A_saida_Y_negative, label='Accel -Y Compensado', linestyle='--', color='red')
plt.title('Accel -Y')
plt.xlabel('Amostra')
plt.ylabel('Valor de Acelerao')
plt.legend()

plt.subplot(3, 2, 5)
plt.plot(amostra, Accel_Z_pos, label='Accel +Z Original', linestyle='-', color='blue')
plt.plot(amostra, A_saida_Z_positive, label='Accel +Z Compensado', linestyle='--', color='red')
plt.title('Accel +Z')
plt.xlabel('Amostra')
plt.ylabel('Valor de Acelerao')
plt.legend()

plt.subplot(3, 2, 6)
plt.plot(amostra, Accel_Z_neg, label='Accel -Z Original', linestyle='-', color='blue')
plt.plot(amostra, A_saida_Z_negative, label='Accel -Z Compensado', linestyle='--', color='red')
plt.title('Accel -Z')
plt.xlabel('Amostra')
plt.ylabel('Valor de Acelerao')
plt.legend()

plt.tight_layout()
plt.savefig('comparacao_graficos_compensados.jpg', format='jpeg', dpi=300, bbox_inches='tight')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np

# Calculando o RMSE para cada par de dados
RMSE_X_pos = np.sqrt(mean_squared_error(Accel_X_pos, A_saida_X_positive))
RMSE_X_neg = np.sqrt(mean_squared_error(Accel_X_neg, A_saida_X_negative))

RMSE_Y_pos = np.sqrt(mean_squared_error(Accel_Y_pos, A_saida_Y_positive))
RMSE_Y_neg = np.sqrt(mean_squared_error(Accel_Y_neg, A_saida_Y_negative))

RMSE_Z_pos = np.sqrt(mean_squared_error(Accel_Z_pos, A_saida_Z_positive))
RMSE_Z_neg = np.sqrt(mean_squared_error(Accel_Z_neg, A_saida_Z_negative))

# Criando uma tabela para os resultados
data = {
    'Eixo': ['X', 'Y', 'Z'],
    'RMSE sem compensao +': [RMSE_X_pos, RMSE_Y_pos, RMSE_Z_pos],
    'RMSE sem compensao -': [RMSE_X_neg, RMSE_Y_neg, RMSE_Z_neg]
}

df = pd.DataFrame(data)
print(df)
```
